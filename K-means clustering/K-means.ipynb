```python
D=dataset.iloc[:, 2:15]
D.head()
```

```python
values_of_data = dataset.iloc[:, 2:15].values
```

```python
no_of_row=values_of_data.shape[0] 
no_of_column=values_of_data.shape[1] 
```

```python
K=3 # number of clusters
```

```python
import numpy as np
C=np.array([]).reshape(no_of_column,0) 
```

```python
import random as rd
for i in range(K):
    random=rd.randint(0,no_of_row-1)
    C=np.c_[C,values_of_data[random]]
C[:,1]
```

```python
iteration=600
```

```python

for i in range(iteration):
    E=np.array([]).reshape(no_of_row,0)
    for k in range(K):
        #tempDist=np.sum((values_of_data-Centroids[:,k])**2,avalues_of_datais=1)
        temp=(D[['Alcohol','Malic.acid','Ash','Acl', 'Mg', 
          'Phenols','Flavanoids','Nonflavanoid.phenols','Proanth','Color.int','Hue','OD','Proline']] - np.array(C[:,k])).pow(2).sum(1).pow(0.5)
        E=np.c_[E,temp]
    cluster_index=np.argmin(E,axis=1)+1
    
    
    cluster={}
    for clus in range(K):
        cluster[clus+1]=np.array([]).reshape(13,0)
    for i in range(no_of_row):
        cluster[cluster_index[i]]=np.c_[cluster[cluster_index[i]],values_of_data[i]]
     
    for clus in range(K):
        cluster[clus+1]=cluster[clus+1].T
    
    for clus in range(K):
        C[:,clus]=np.mean(cluster[clus+1],axis=0)  
```

```python
C=C.T
color=['red','blue','green']
labels=['1st_cluster','2nd_cluster','3rd_cluster']
for k in range(3):
    plt.scatter(cluster[k+1][:,-1],cluster[k+1][:,1],c=color[k],label=labels[k])
plt.scatter(C[:,-1],C[:,1],s=200,c='yellow',label='Centroids')
plt.legend()
plt.show()
```
