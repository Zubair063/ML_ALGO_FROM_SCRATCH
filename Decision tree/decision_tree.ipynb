```python
def get_target_row_count(rows):
    target_row_count = {}  
    for row in rows:
        label = row[-1]
        if label not in target_row_count:
            target_row_count[label] = 0
        target_row_count[label] += 1
    return target_row_count
```

```python
def find_numeric(value):
    return isinstance(value, int) or isinstance(value, float)
```

```python
class ques_to_find_node:

    def __init__(self, col_val, v):
        self.col_val = col_val
        self.v = v

    def matching_function(self, e):
        value = e[self.col_val]
        if find_numeric(value):
            return value >= self.v
        else:
            return value == self.v

    def __repr__(self):
        con = "=="
        if find_numeric(self.v):
            con = ">="
        return "Is %s %s %s?" % (
            head[self.col_val], con, str(self.v))
```

```python
def partitioning_tree(rows, question):
    positive, negative = [], []
    for row in rows:
        if question.matching_function(row):
            positive.append(row)
        else:
            negative.append(row)
    return positive, negative
```

```python
def index_gini(rows):
    target_row_count = get_target_row_count(rows)
    impure = 1
    for lbl in target_row_count:
        prob_of_lbl = target_row_count[lbl] / float(len(rows))
        impure -= prob_of_lbl**2
    return impure
```

```python
def gaining_information(true, false, total_uncertanity):
    p = float(len(true)) / (len(true) + len(true))
    return total_uncertanity - p * index_gini(true) - (1 - p) * index_gini(false)
```

```python
def splitting (rows):
    
    chosen_info_gain = 0  
    chosen_question = None  
    total_uncertanity = index_gini(rows)
    no_of_features = len(rows[0]) - 1  

    for col in range(no_of_features):  

        values = set([row[col] for row in rows])  
        for val in values:  

            question = ques_to_find_node(col, val)

           
            positive, negative = partitioning_tree(rows, question)

            if len(positive) == 0 or len(negative) == 0:
                continue

            gain = gaining_information(positive, negative, total_uncertanity)
            if gain >= chosen_info_gain:
                chosen_info_gain, chosen_question = gain, question

    return chosen_info_gain, chosen_question
```

```python
class leaf_node:
   
    def __init__(self, rows):
        self.predict = get_target_row_count(rows)
```

```python
class nodes:
    
    def __init__(self,
                 question,
                 positive,
                 negative):
        self.question = question
        self.positive = positive
        self.negative = negative
```

```python
def tree(rows):
   

    g, q = splitting (rows)

  
    if g == 0:
        return leaf_node(rows)

    
    positive, negative = partitioning_tree(rows, q)

    
    true = tree(positive)

    
    false = tree(negative)
    return nodes(q, true, false)
```

```python
def print_the_tree(enter_node, spacing=""):
    
    if isinstance(enter_node, leaf_node):
        print (spacing + "Predict", enter_node.predict)
        return

    
    print (spacing + str(enter_node.question))

    print (spacing + '--> True:')
    print_the_tree(enter_node.positive, spacing + "  ")

    print (spacing + '--> False:')
    print_the_tree(enter_node.negative, spacing + "  ")
```

```python
my_tree = tree(traning_list)
```

```python
print_the_tree(my_tree)
```

```python
def classification(row, node):
    if isinstance(node, leaf_node):
        return node.predict

    if node.question.matching_function(row):
        return classify(row, node.positive)
    else:
        return classify(row, node.negative)
```

```python
def print_the_leaf_node(get_target_row_count):
    t = sum(get_target_row_count.values()) * 1.0
    p = {}
    for l in get_target_row_count.keys():
        p[l] = str(int(get_target_row_count[l] / t * 100)) + "%"
    return p
```
